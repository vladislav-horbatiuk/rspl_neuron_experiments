```python
%matplotlib widget
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10, 10, 0.1)
sigmoid = 1 / (1 + np.exp(-x))
step = np.where(x >= 0, 1.0, 0.0)
plt.figure()
plt.plot(x, sigmoid, label='sigmoid', linewidth=3)
plt.plot(x, step, label='step', linewidth=3)
plt.axhline(color='black')
plt.axvline(color='black')
plt.legend()
plt.show()
```

```python
# Make data.
Sm = np.arange(-5, 5, 0.1)
Sh = np.arange(-5, 5, 0.1)
Sm, Sh = np.meshgrid(Sm, Sh)

Zact = (5*Sm) / (1 + np.exp(-Sh)) + (Sm) / (1 + np.exp(Sh))
Zpdwp = (Sm) / (1 + np.exp(-Sh))
Zpdwm = (Sm) / (1 + np.exp(Sh))
Zpdh = (Sm) / (2 + np.exp(-Sh) + np.exp(Sh))
```

```python
ax = plt.figure().add_subplot(projection='3d')
surf = ax.plot_surface(Sm, Sh, Zact, cmap=plt.cm.gist_heat)
surf._edgecolors2d = surf._edgecolor3d
surf._facecolors2d = surf._facecolor3d
ax.set_zlim(-20, 20)
ax.set_xlabel(r'$S_m$')
ax.set_ylabel(r'$S_h$')
ax.set_zlabel(r'$a(5*S_m, S_m, S_h)$')

plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')

surf1 = ax.plot_surface(Sm, Sh, Zpdwp, color='red', facecolor='red',
                        label=r'$\frac{\partial f}{\partial w_{+(q)}}$', alpha=0.7)
surf1._edgecolors2d = surf1._edgecolor3d
surf1._facecolors2d = surf1._facecolor3d
surf2 = ax.plot_surface(Sm, Sh, Zpdwm, color='green', facecolor='green',
                        label=r'$\frac{\partial f}{\partial w_{-(q)}}$', alpha=0.7)
surf2._edgecolors2d = surf2._edgecolor3d
surf2._facecolors2d = surf2._facecolor3d
ax.set_zlim(-3, 3)
ax.set_xlabel(r'$x_q$')
ax.set_ylabel(r'$s_h$')
ax.set_zlabel(r'$\frac{\partial f}{\partial w_{+(q)}},\frac{\partial f}{\partial w_{-(q)}}$')
ax.legend()

plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')

surf3 = ax.plot_surface(Sm, Sh, Zpdh, cmap=plt.cm.gist_heat)
surf3._edgecolors2d = surf3._edgecolor3d
surf3._facecolors2d = surf3._facecolor3d

# Customize the z axis.
ax.set_zlim(-1.5, 1.5)
ax.set_xlabel(r'$x_q$')
ax.set_ylabel(r'$s_h$')
ax.set_zlabel(r'$\frac{\partial f}{\partial h_q}$')
ax.legend()

#ax.zaxis.set_major_locator(LinearLocator(6))

plt.show()
```
